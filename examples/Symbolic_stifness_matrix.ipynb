```python
L = 10.0
no_of_element=1
le_val = L/no_of_element
E = 10000
b = 1
t = 1
nu = 0.0
alpha = 5/6
EI_val =  E*b*t**3/12
G = E/(2*(1+nu))
GA_val = alpha*G*b*t

```

```python
#eigenvalues for Mixed order elements
FEM_21_eig = FEM_21()[0].subs({le: le_val, EI: EI_val,GA :GA_val}).eigenvals()
print("eigenvalues for Mixed order elements FEM:")
display(FEM_21_eig)
VEM_21_eig = VEM_21()[0].subs({le: le_val, EI: EI_val,GA :GA_val}).eigenvals()
print("eigenvalues for Mixed order elements VEM:")
display(VEM_21_eig)
```

```python
#eigenvalues for 2nd order elements
FEM_22_eig = FEM_22()[0].subs({le: le_val, EI: EI_val,GA :GA_val}).eigenvals()
print("eigenvalues for 2nd order elements FEM:")
display(FEM_22_eig)
VEM_22_eig = VEM_22()[0].subs({le: le_val, EI: EI_val,GA :GA_val}).eigenvals()
print("eigenvalues for 2nd order elements VEM:")
display(VEM_22_eig)
```
